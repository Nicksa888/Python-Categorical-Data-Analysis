# The “goodness of fit” test statistic

At this point, we have our observed frequencies and a collection of probabilities corresponding the null hypothesis that we 
want to test. We’ve stored these in Python as the corresponding variables observed and probabilities. What we want to do now is 
construct a test of the null hypothesis. We’re going to need a test statistic. 
The basic trick that a goodness of fit test uses is to construct a test statistic that measures how “close” the data are to 
the null hypothesis. If the data don’t resemble what you’d “expect” to see if the null hypothesis were true, then it probably
isn’t true. Okay, if the null hypothesis were true, what would we expect to see? Or, to use the correct terminology, what are 
the **expected frequencies*?. There are 56578 observations, and (if the null is true) the probability of any one of them 
choosing explosives weapon being chosen is 56578 * .25 = 14144.5, so I guess we’re expecting counts of explosives weapon, right? 
Lets calculate it:

```python
# We have a sample size of 5342, so to calculate the expected weapon type use count is:

N = 56578 #sample size
expected = N * probabilities
expected
```

Now, how do we translate this into a test statistic? Clearly, what we want to do is compare the expected number of observations 
in each category with the observed number of observations in that category. And on the basis of this comparison, we ought to be 
able to come up with a good test statistic. To start with, let’s calculate the difference between what the null hypothesis 
expected us to find and what we actually did find.

```python
observed-expected
```

Regardless of whether we do the calculations by hand or whether we do them in Python, it’s clear that more explosives were used
and fewer incendiary weapons than the null hypothesis predicted. However, a moment’s thought suggests that these raw differences
aren’t quite what we’re looking for. Intuitively, it feels like it’s just as bad when the null hypothesis predicts too few 
observations (which is what happened with explosives) as it is when it predicts too many (which is what happened with incendiary
weapons). So it’s a bit weird that we have a negative number for clubs and a positive number for hearts. One easy way to fix 
this is to square everything, so that we now calculate the squared differences. As before, we could do this by hand, but it’s 
easier to do it in Python…

```python
(observed - expected)**2
```

```python
# If we convert these into actual numbers, they are as follows:
175,390,300
16,789,510
35,802,270
128,992,800
# Therefore, what we’ve got now is a collection of numbers that are significantly higher when the null hypothesis makes a bad 
#prediction (explosives and incendiary), but are small whenever it makes a good one (firearms and other weapon)
```

```python
# Next, let’s also divide all these numbers by the expected frequency:
(observed - expected)**2 / expected
```

In effect, what we’ve got here are four different “error” scores, each one telling us how big a “mistake” the null hypothesis 
made when we tried to use it to predict our observed frequencies. So, in order to convert this into a useful test statistic, 
one thing we could do is just add these numbers up. The result is called the goodness of fit statistic, conventionally referred 
to either as goodness of fitness test or GOF. We can calculate it using this command in Python:

```python
sum((observed - expected)**2/expected)
```

Intuitively, it’s clear that if the gof is small, then the observed data are very close to what the null hypothesis predicted, 
so we’re going to need a large statistic in order to reject the null. As we’ve seen from our calculations, in our data set 
we’ve got a value of 25237.716214783133. So now the question becomes, is this a big enough value to reject the null?

The final step in the process of constructing our hypothesis test is to figure out what the rejection region is. That is, what 
values of the test statistic would lead us to reject the null hypothesis. As we saw earlier, large values imply that the null 
hypothesis has done a poor job of predicting the data from our experiment, whereas small values of imply that it’s actually 
done pretty well. Therefore, a pretty sensible strategy would be to say there is some critical value, such that if the value is 
bigger than the critical value we reject the null; but if the value is smaller than this value we retain the null. Right, so all 
we have to do is figure out what this critical value is. And it’s pretty straightforward. If we want our test to have 
significance level of 95% (that is, we are willing to tolerate a Type I error rate of 5%), then we have to choose our critical 
value so that there is only a 5% chance that the value could get to be that big if the null hypothesis is true. That is to say, 
we want the 95th percentile of the sampling distribution. The 95th percentile of a chi-squared distribution with 3 degrees of 
freedom can be calculated like this by using chi2.ppf() in python:

```python
from scipy.stats import chi2
round(chi2.ppf(0.95, 3), 2)
```

Therefore, since our value is 25237.716214783133, we can reject the null hypothesis and conclude that weapon choices are not 
chosen randomly. If we want an exact value, we can calculate it using the chisquare() function from scipy.stats:

```python
from scipy.stats import chisquare

chisquare(f_obs = observed, f_exp = expected)
```

A p-value of 0.0 suggests very strong evidence against the null hypothesis.

# The chi square test of association or independence

```python
# First we subset the weapon type and group variables:
selected_columns = ['Group', 'Weapon']
SA_data_1 = SA_data[selected_columns]
```

```python
observations = pd.crosstab(index = SA_data_1["Weapon"], columns = SA_data_1["Group"], margins = False)
observations
```

```python
# Our hypothesis is that groups attack using different weapons. 
```

```python
# Now that we have a frequency table, we can use the chi2_contingency function from scipy.stats to crunch the numbers:

from scipy.stats import chi2_contingency

chi2, p, dof, ex = chi2_contingency(observations, correction = True) # Yates correction
print("chi2 = ", chi2)
print("p = ", p)
print("degrees of freedom = ", dof)
print("expected = ", ex)
```

```python
chi2 is the test statistic, p is the p-value, dof is the degrees of freedom for the test, and ex tells you the “expected”
values, that is, the frequency counts you would expect to see under the null hypothesis. 
The null hypothesis, by the way, is that the variables are independent of one another, that is, that the choice of weapon has 
nothing to do with whether the group is Taliban or OtherGroup. The alternative, of course, is that they are not independent;
that is, that each group will tend to make different choices. Or, put another way, that the frequency of each weapon type
chosen is contingent on whether or not the group is Taliban or OtherGroup.
```

# Effect Size

```python
So, let’s suppose that you’ve run your chi-square test, which turns out to be significant. So you now know that there is some 
association between your variables (independence test) or some deviation from the specified probabilities (goodness of fit 
test). Now you want to report a measure of effect size. That is, given that there is an association/deviation, how strong is 
it? There are several different measures that you can choose to report, and several different tools that you can use to 
calculate them. Pingouin provides you with one automatically: Cramer’s V. The idea here is that the cramers v statistic is 
    supposed to range between 0 (no at all association) and 1 (perfect association).
```

```python
import pingouin as pg
expected, observed_pingouin, stats = pg.chi2_independence(SA_data, x = 'Weapon', y = 'Group')
stats
```

```python
# The cramers v figures all are found between 0.20 and 0.30, so moderate effect size.
```
